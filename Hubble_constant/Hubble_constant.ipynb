```python
from astropy.io import ascii

data = ascii.read("Data_Cosmology_Class/Union2.txt")
```

```python
print(data)
```

```python
import pylab as pl
import numpy as np

# When Dl is in Mpc
# mu = 5.0*log (Dl) + 25.0

pl.scatter(10**( (data["mu"] - 25.0)/5.0), 3.e5*data["z"])
pl.xlim(0, 150)
pl.ylim(0, 10000)

xx = np.arange(400)

pl.plot(xx, 65.*xx)
pl.plot(xx, 70.*xx)
pl.plot(xx, 75.*xx)
pl.xlabel("v (km/s/Mpc)")
pl.ylabel("D (Mpc)")
```
